```python
import numpy as np
from mygame_functions import GameSimulationGC
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
M=5
Ns=127
Np=430
T=100000
ponderas=1
imprime=1
meangains=np.zeros(T)
meangainp=np.zeros(T)
for i in range(ponderas):
    A, GS, GP, nplayers = GameSimulationGC(Ns, Np, M, T)
    for t in range(T):
        if t<1000:
            meangains[t] += np.mean(GS[:(t+1)])
            meangainp[t] += np.mean(GP[:(t+1)])
        else:
            meangains[t] += np.mean(GS[(t-999):t])
            meangainp[t] += np.mean(GP[(t-999):t])
    if (i+1)%imprime==0:
        print(' - Pondera nº{}/{}'.format(i+1,ponderas))
        print('          <gs(T)> = {}'.format(meangains[-1]/(i+1)))
        print('          <gp(T)> = {}'.format(meangainp[-1]/(i+1)))
meangains = meangains/ponderas
meangainp = meangainp/ponderas
```

```python
fig, ax = plt.subplots(1, figsize=(12,6))
ax.set_xscale('log')
#ax.set_yscale('log')
#ax.set_ylim([-5,-1])

ax.set_xlabel(r'$t$', fontsize='large')
ax.set_ylabel(r'$G_i$', fontsize='large')

ax.tick_params(which='minor', direction = 'inout', length=3, width=1,
              labelbottom=False, labeltop=False, labelleft=False, labelright=False,
                  bottom=True, top=True, left=True, right=True)

ax.tick_params(which='major', direction = 'inout', length=6, width = 2,
                  labelbottom=True, labeltop=False, labelleft=True, labelright=False,
                  bottom=True, top=True, left=True, right=True)

cols = ['dodgerblue', 'yellowgreen', 'crimson', 'forestgreen', 'cyan', 'darkblue']

lin12, = ax.plot(np.arange(10,T-1), meangainp[10:-1], lw=1, ls = '-', color = cols[3], alpha=.8)
lin101, = ax.plot(np.arange(10,T-1), meangains[10:-1], lw=1, ls = '-', color = 'k', alpha=.8)

labels = [r'$G_{P}$', r'$G_{S}$']
ax.legend([(lin12,), (lin101,) ], labels, loc='center', bbox_to_anchor=[1.1,0.5])
```

```python
M=5
PNP = np.linspace(1, 30, 8)
Nprods = PNP * 2**M
Nprods = Nprods.astype(int)
index_odd = np.argwhere(Nprods % 2 != 0).flatten()
Nprods[index_odd] = Nprods[index_odd] + 1
N_NP = Nprods / 2**M
```

```python
Nprods
```

```python
M = 5
Ns = 127
sigma = []
GS = []
GP = []
NPS = []
T = 7500
numponderas = 750
imprimecada = 75
for Np in Nprods:
    print('Nprods = {}        ({}/{})'.format(Np, list(Nprods).index(Np) + 1, len(Nprods)))
    s = 0
    ys = 0
    yp = 0
    nps = 0
    for i in range(numponderas):
        A, meangain_s, meangain_p, nplayers_s = GameSimulationGC(Ns, Np, M, T)
        s += np.var(A)/(Ns+Np)
        ys += np.mean(meangain_s)
        yp += np.mean(meangain_p)
        nps += np.mean(nplayers_s)
        if (i + 1) % imprimecada == 0:
            print('   Ponderacion nº {}/{}'.format(i + 1, numponderas))
            print('     Nprods = {}        ({}/{})'.format(Np, list(Nprods).index(Np) + 1, len(Nprods)))
            print('  effec alpha = {:.3f}'.format(2**M/np.mean(nplayers_s)))
            print('         σ²/N = {:.3f}'.format(s / (i + 1)))
            print('     <gain_s> = {:.3f}'.format(ys / (i + 1)))
            print('     <gain_p> = {:.4f}'.format(yp / (i + 1)))
            print(' <Nplayers_s> = {:.2f}'.format(nps / (i + 1)))
            print(' ')
    sigma.append(s / numponderas)
    GS.append(ys / numponderas)
    GP.append(yp / numponderas)
    NPS.append(nps / numponderas)
```
