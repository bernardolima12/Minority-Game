```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
from mygame_functions import GameSimulationSPriv
```

```python
Ns = 81
Np = 71
M = 7
numponderas = 300
imprimecada = 30
used_strategies = []
gain_priv = []
esesprimas = []
SP = 1
rango_s = 26
for eses in range(rango_s):
    if eses < 10:
        SP += 1
    else:
        SP += 2
    esesprimas.append(SP)
    print("        ----------------------")
    print("        S' = {}        ({}/{})".format(SP, eses + 1, rango_s))
    print("        ----------------------")
    g = 0
    us = 0
    for i in range(numponderas):
        gain, numstrats = GameSimulationSPriv(Ns, Np, M, SP, T=10000)
        g += np.mean(gain)
        us += numstrats
        if (i + 1) % imprimecada == 0:
            print('      ---> Ponderacion nÂº {}/{}'.format(i + 1, numponderas))
            print("      S' = {}        ({}/{})".format(SP, eses + 1, rango_s))
            print('            <gain> = {}'.format(g / (i + 1)))
            print('          <s_used> = {}'.format(us / (i + 1)))
            print(' ')
    
    gain_priv.append(g / numponderas)
    used_strategies.append(us / numponderas)
    print('Remembering')
    print("S'              = ", esesprimas)
    print('Gs              = ', gain_priv)
    print('Used strategies = ', used_strategies)
```

```python
file_name = "PriviS_Sused_gainVIRTUALbien_" + str(Ns) + "Nspecs_" + str(
    Np) + 'Nprods_' + str(M) + 'm_' + str(numponderas) + "numponderas"
with open('{}.dat'.format(file_name), 'w') as file:
    for i in range(len(esesprimas)):
        file.write('{} {} {}\n'.format(esesprimas[i], used_strategies[i],
                                       gain_priv[i]))
```
