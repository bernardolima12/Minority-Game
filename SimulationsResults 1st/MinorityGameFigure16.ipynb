```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
#import sys
```

```python
%matplotlib inline
```

```python
numponderas = 100
file_name = "deltaM__infos_difgains_" + str(numponderas) + "ponderas"
```

```python
mpr = []
HMP = []
GMP = []
with open('{}.dat'.format(file_name), 'r') as file:
    for line in file:
        value = line.split()
        value[2].split('\n')
        mpr.append(float(value[0]))
        HMP.append(float(value[1]))
        GMP.append(float(value[2]))
```

```python
fig, ax = plt.subplots(2, sharex=True,gridspec_kw={'hspace': 0},figsize=(10, 8))

ax[0].tick_params(which='both',labelbottom=False,labeltop=True,labelleft=True,
                  labelright=False,bottom=True,top=True,left=True,right=True)
ax[1].tick_params(which='both',labelbottom=True,labeltop=False,labelleft=True,
                  labelright=False,bottom=True,top=True,left=True,right=True)

#ax_top = ax[1].twiny()
#ax_top.tick_params(direction='inout', labelbottom=False,labeltop=False,labelleft=False,
#                  labelright=False)

ax[0].plot(mpr, HMP, clip_on=True, marker='o', mec='b', mfc='w', ls='-', color='k', alpha = 0.8)
ax[0].set_ylabel(r"$H_{M'}/N$")

ax[1].set_ylabel(r"$\Delta$G = $G_{M'}-G_{M}$")
ax[1].set_xlabel(r"$\Delta$ M = M'-M")
ax[1].plot(mpr, GMP, clip_on=True, marker='o', mec='b', mfc='w', ls='-', color='k', alpha = 0.8)
fig.savefig('Hmprima,gain_vs_Mprima__alfa01,M3_100pond10000T_emptiesuman.png')
```
