```python
import numpy as np
import matplotlib.pyplot as plt
from mygame_functions import one_simulation_info
```

```python
%matplotlib inline
```

```python
S = 2
```

```python
M = 8
```

```python
alpha = [10**(-0.65 + 0.145*ni)  for ni in range(13)]  #Logaritmically equispaced
alpha
```

```python
enes = [int(2**M/alpha[i]) for i in range(len(alpha))]  # N has to be integer
enes
```

```python
alpha = [2**M/enes[i] for i in range(len(alpha))]  # X
alpha
```

```python
num_ponderas = 50
theta = []
```

```python
T = 10000
y = []
for j in range(num_ponderas):
    print('Ponderacion nº {}/{}'.format(j+1, num_ponderas))
    times, attendances, information = one_simulation_info(enes[0], S, M, T)
    y.append(np.mean(information))
    print('<H> = ', y[-1])
```

```python
theta.append(np.mean(y))
```

```python
theta
```

```python
num_ponderas = 2
T = 100000
y = []
for j in range(num_ponderas):
    print('Ponderacion nº {}/{}'.format(j+1, num_ponderas))
    times, attendances, information = one_simulation_info(enes[0], S, M, T, imprime=1000)
    y.append(np.mean(information))
    print('<H> = ', y[-1])
```

```python
num_ponderas = [20,40,60,80,100,150,180,210,240,270,300,300]
T = 10000
k=0
for N in enes[1:]:
    print('N={} ({}/{})'.format(N, enes.index(N)+1, len(enes)))
    print(r'alpha = {:.3f}'.format(alpha[enes.index(N)]))
    y = []
    for j in range(num_ponderas[k]):
        print('     Ponderacion nº {}/{}'.format(j+1, num_ponderas[k]))
        times, attendances, information = one_simulation_info(N, S, M, T)
        y.append(np.mean(information))
    k+=1
    theta.append(np.mean(y))
    print('<H> = ', theta[-1])
```

```python
enes
```

```python
file_name = "theta_vs_alpha_jupyter"+str(M)+"m"
with open('Data/{}.dat'.format(file_name), 'w') as file:
    for i in range(len(theta)):
        file.write('{} {}\n'.format(alpha[i], theta[i]))
```

```python
alpha8 = []
theta8 = []
M = 8
file_name = "theta_vs_alpha_jupyter"+str(M)+"m"
with open('Data/{}.dat'.format(file_name), 'r') as file:
    for line in file:
        value = line.split()
        value[1].split('\n')
        alpha8.append(float(value[0]))
        theta8.append(float(value[1]))
del theta8[-1]
del theta8[-1]
del alpha8[-1]
del alpha8[-1]
```

```python
alpha5 = []
theta5 = []
M = 5
file_name = "theta_vs_alpha_jupyter"+str(M)+"m"
with open('Data/{}.dat'.format(file_name), 'r') as file:
    for line in file:
        value = line.split()
        value[1].split('\n')
        alpha5.append(float(value[0]))
        theta5.append(float(value[1]))
```

```python
from matplotlib import ticker
fig, ax = plt.subplots()
ax.set_xscale('log')
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel('H')
ax.set_xlim(0.11,12)
ax.set_ylim(0,0.4)
ax.plot(alpha5, theta5, '.', clip_on=False, label = 'm = 5')
ax.plot(alpha8, theta8, '.', clip_on=False, label = 'm = 8')
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: (
    '{{:.{:1d}f}}'.format(int(np.maximum(-np.log10(x), 0)))).format(x)))
ax.set_yticks([0.00,0.10,0.20,0.30,0.40])
ax.legend()
fig.savefig("HvsAlpha_m=5y8.png")
```
