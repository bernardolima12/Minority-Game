```python
import numpy as np
from mygame_functions import GameSimulationInfo
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
M = 7
alpha = [10**(-0.9 + 0.06*ni)  for ni in range(28)]  
enes = np.array([int(2**M/alpha[i]) for i in range(len(alpha))])  # N has to be integer
for n in range(len(enes)):
    if enes[n]%2==0:
        enes[n]+=1
enes
```

```python
alpha = np.array([2**M/enes[i] for i in range(len(alpha))])
alpha
```

```python
plt.plot(alpha, np.zeros(len(alpha)), '.')
plt.xscale('log')
```

```python
M = 7
H = []
H2 = []
numponderas = 100
imprimecada = 10
for N in enes:
    print('    N = {}    ({}/{})'.format(N,len(H)+1,len(enes)))
    print('--> α = {:.4f} <-- \n'.format(alpha[len(H)]))
    h = 0
    h2 = 0
    if N>=509:
        T = 10000
    if N<509:
        T = 50000
    if N<86:
        T = 75000
    if N<44:
        T = 90000
    
    for i in range(numponderas):
        info,info2 = GameSimulationInfo(N, M, T)
        h += info
        h2 += info2
        if (i+1)%imprimecada == 0:
            print('   Ponderacion nº {} de {}'.format(i+1,numponderas))
            print('      -H_conlen = {:.5f}'.format(h/(i+1)))
            print('      -H_sinlen = {:.5f}'.format(h2/(i+1)))
        
    H.append(h/numponderas)
    H2.append(h2/numponderas)
    
file_name = 'HvsALPHA_' + '7M_' + str(numponderas) + 'numponds'
with open('{}.dat'.format(file_name), 'w') as file:
    for i in range(len(alpha)):
        file.write('{} {} {}\n'.format(alpha[i], H[i], H2[i]))
```

```python
plt.plot(alpha,H,'.',label = 'pongo len 0')
plt.plot(alpha,H2,'.',label = 'quito len 0')
plt.xscale('log')
plt.legend(loc=0)
```
