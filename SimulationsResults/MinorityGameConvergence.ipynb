```python
import numpy as np
import matplotlib.pyplot as plt
from mygame_functions import one_simulation, one_simulation_variables
%matplotlib inline
```

```python
S=2
```

```python
N = 199
M = 5
```

```python
times, attendances, meanA = one_simulation(N, S, M, T=5000, imprime = 500)
```

```python
fig, ax = plt.subplots(figsize = (12,6))

ax.axhline(y=0, color='k', linestyle='--')
ax.plot(times, meanA, '-', label = r'$<A(t)>$')
ax.set_xlabel('t')
ax.set_ylabel('<A(t)>')

axin = fig.add_axes([0.27,0.55,0.6,0.3])
axin.axhline(y=0, color='k', linestyle='--')
axin.plot(times,meanA, '-')
axin.set_xlabel('t')
axin.set_ylabel('<A(t)>')
desde = 500
hasta = 5000
axin.set_xlim([desde,hasta])
axin.set_ylim([-1,1])
axin.set_xticks(range(desde,hasta+1,1500))
```

```python
filename = '<A>_N={}_M={}.png'.format(N,M)
fig.savefig(filename)
```
