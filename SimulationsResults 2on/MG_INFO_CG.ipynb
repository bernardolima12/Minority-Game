```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.ticker import MultipleLocator
from MinorityGameComplex import GameSimulationImitationTermalINFO
```

```python
S=2
N=401
T=25000
dynT=2500
numponderas = 5
imprimecada = 1

lenemes = 10
emes = np.arange(3,3+lenemes)
alpha=2**emes/N

nimis = np.array([1,200,300,400])
lenimis = len(nimis)

sigma = np.zeros((lenimis, lenemes))
INFOS = np.zeros((lenimis, lenemes))

for m in range(lenemes):
    M = emes[m]
    for ncop in range(lenimis):
        Ncop = nimis[ncop]
        print(' ..............................')
        print('              M = {}     ({} / {})'.format(M, m+1, lenemes))
        print('              α = {}'.format(alpha[m]))
        print('           Ncop = {}     ({} / {})'.format(Ncop, ncop+1, lenimis))
        print(' ..............................')
        s = 0
        ifo = 0
        for pond in range(numponderas):
            A, I = GameSimulationImitationTermalINFO(S, M, N, Ncop, T, dynT)
            s += np.var(A) / N
            ifo += np.mean(I) / N
            if (pond + 1) % imprimecada == 0:
                print('        -Ponderacion nº {}/{}'.format(pond + 1, numponderas))
                print('             σ²/N = {}'.format(s / (pond + 1)))
                print('              H/N = {}'.format(ifo / (pond + 1)))
        sigma[ncop,m] =  s / numponderas
        INFOS[ncop,m] =  ifo / numponderas
        print('...Remembering...')
        print('Ncops = ', nimis[:(ncop+1)])
        print('σ²/N = ', sigma[:(ncop+1),m])
        print(' H/N = ', INFOS[:(ncop+1),m])
```

```python
fig, ax = plt.subplots(1, figsize=(8,6))
ax.set_xscale('log')
#ax.set_yscale('log')

ax.set_xlabel(r'$\alpha$', fontsize='large')
#ax.set_ylabel(r'$\sigma^2/N$', fontsize='large')

ax.tick_params(which='minor', direction = 'inout', length=3, width=1,
              labelbottom=False, labeltop=False, labelleft=False, labelright=False,
                  bottom=True, top=True, left=True, right=True)

ax.tick_params(which='major', direction = 'inout', length=6, width = 2,
                  labelbottom=True, labeltop=False, labelleft=True, labelright=False,
                  bottom=True, top=True, left=True, right=True)

cols = ['k', 'dodgerblue', 'yellow', 'crimson', 'forestgreen']

lin = ['1','2','3','4','5']
mark = ['1','2','3','4','5']
for i in range(len(nimis)):
    nc=i
    lin[nc], = ax.plot(alpha[1:], INFOS[nc][1:], lw=0.5, ls = '--', color = 'k', alpha=0.5)
    mark[nc], = ax.plot(alpha[1:], INFOS[nc][1:], ls = '', marker = 'o', mec='k',mfc=cols[nc],
                     ms = 7, mew = 1, alpha=0.9)



lin[0], = ax.plot(alpha[1:], INFOS[0][1:], lw=1, ls = '-', color = 'k', alpha=0.8)
mark[0], = ax.plot(alpha[1:], INFOS[0][1:], ls = '', marker = 'o', mec='k',mfc=cols[0],
                 ms = 7, mew = 1, alpha=0.9)
    
"""yticks1 = [1]
ax.set_yticks(yticks1)
ax.set_yticklabels([str(i) for i in yticks1])"""

xticks1 = [0.1, 1, 10]
ax.set_xticks(xticks1)
ax.set_xticklabels([str(i) for i in xticks1])

ax.axhline(y=0, xmin=0, xmax=1, color='k', alpha=0.7, linestyle='--')

labels = [(r'$N_I = $ ' + str(nimis[i])) for i in range(len(nimis))]
ax.legend([(lin[i], mark[i]) for i in range(len(lin))], 
          labels, loc='upper center', bbox_to_anchor=[0.4,1])

filename = 'CG_INFOvsALPHA' + str(
                    N) +'N_' + 'Ncop_CG_' + str(
                    dynT) + 'dynT_' + str(T)+'T_'+str(numponderas)+'numponds.png'
fig.savefig(filename, bbox_inches='tight')
```
