```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from MinorityGameComplex import GameSimulationImitationSW
```

# Change with alpha for a couple situations of $N_{imitators}$

```python
S=2
emes = [(3 + i) for i in range(8)]
N=201
Ncop = 100
p=0.5

numponderas = 10
imprimecada = 5

alpha = []
sigma = []
gains = []
gains_imitators = []
for M in emes:
    alpha.append(2**M / N)
    print(' ...................')
    print(' M = {}        {} / {}'.format(M, emes.index(M)+1, len(emes)))
    print(' Œ± = {:.3f}'.format(alpha[-1]))
    print(' ...................')
    s = 0
    g = 0
    gc = 0
    for pond in range(numponderas):
        A, G, GC = GameSimulationImitationSW(S, M, N, Ncop, p, T=10000)
        s += np.var(A) / N
        g += np.mean(G)
        gc += np.mean(GC)
        if (pond + 1) % imprimecada == 0:
            print('        -Ponderacion n¬∫ {}/{}'.format(pond + 1, numponderas))
            print('        Œ± = {:.3f}'.format(alpha[-1]))
            print('             œÉ¬≤/N = {}'.format(s / (pond + 1)))
            print('             <g>  = {}'.format(g / (pond + 1)))
            print('             <gc> = {}'.format(gc / (pond + 1)))
    sigma.append(s / numponderas)
    gains.append(g / numponderas)
    gains_imitators.append(gc / numponderas)
alpha = np.array(alpha)
sigma = np.array(sigma)
gains = np.array(gains)
gains_imitators = np.array(gains_imitators)
```
